```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df = pd.read_csv("./dataset/housing.csv", delim_whitespace=True, header=None)
'''
print(df.info())
print(df.head())
'''
```

```python
#CRM : 인구 1인당 범죄 발생률
#ZN : 25,000평방피트 이상의 주거 구역 비중
#INDUS : 소매업 외 상업이 차지하는 면적 비율
#CHAS: 찰스강 위치변수(1:강주변, 0:이외)
#NOX: 일산화질소 농도
#RM : 집의 평균 방 수
#AGE : 1940년 이전에 지어진 비율
#DIS : 5가지 보스턴 시 고용 시설까지의 거리
#RAD : 순환고속도로의 접근 용이성
#TAX : $10,000당 부동산 세율 총계
#PIRATIO : 지역별 학생과 교사 비율
#B : 지역별 흑인 비율
#LSTAT : 급여가 낮은 직업에 종사하는 인구 비율(%)
#가격 : (단위 : $1,000)

df
```

```python
dataset = df.values
X = dataset[:,0:13]
Y = dataset[:,13]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train, Y_train, epochs=200, batch_size=10)
```

```python
# 예측 값과 실제 값의 비교
Y_prediction = model.predict(X_test).flatten()
for i in range(10):
    label = Y_test[i]
    prediction = Y_prediction[i]
    print(f"실제가격: {format(label,'.3f')}, 예상가격: {format(prediction,'.3f')}")
```
